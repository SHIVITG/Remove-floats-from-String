```python
import csv
import glob
```

```python
files = glob.glob('TextData/*.txt')
for j in files:
    with  open(j,"r+") as i:
        text= i.readlines()
        print(type(text))
        print("The original list : " + str(text))
        f = "Data/" + j
        print(f)
        file = open(f,"w+")
        res = [] 
        for ele in text: 
            try: 
                float(ele)
            except ValueError: 
                res.append(ele)
        print("String list after filtering floats : " + str(res)) 
        file.write(str(res))
```
